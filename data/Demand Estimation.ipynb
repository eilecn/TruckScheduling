```python
from _ast import Store

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import *
```

```python
Demand_org = pd.read_csv('data/WoolworthsDemand2024.csv')
print(Demand_org['Store'].values)
Location_org = pd.read_csv('data/WoolworthsLocations.csv')
Demand = Demand_org.melt(id_vars=['Store'], var_name='Date', value_name='Pallets')
Demand['Week'] = 'Week 1'
Demand['Week_day'] = 'Monday'
for j in range(64):
    i0  = j*28
    i1  = j*28 + 7
    i2  = j*28 + 14   
    i3  = j*28 + 21
    i4  = j*28 + 28
    Demand.loc[Demand.index[i0 : i1], 'Week'] = 'Week 1'
    Demand.loc[Demand.index[i1 : i2], 'Week'] = 'Week 2'
    Demand.loc[Demand.index[i2 : i3], 'Week'] = 'Week 3'
    Demand.loc[Demand.index[i3 : i4], 'Week'] = 'Week 4'
    for i in range(4):
        Demand.loc[Demand.index[j*28+i*7], 'Week_day'] = '1-Mon'
        Demand.loc[Demand.index[j*28+i*7+1], 'Week_day'] = '2-Tue'
        Demand.loc[Demand.index[j*28+i*7+2], 'Week_day'] = '3-Wed'
        Demand.loc[Demand.index[j*28+i*7+3], 'Week_day'] = '4-Thu'
        Demand.loc[Demand.index[j*28+i*7+4], 'Week_day'] = '5-Fri'
        Demand.loc[Demand.index[j*28+i*7+5], 'Week_day'] = '6-Sat'
        Demand.loc[Demand.index[j*28+i*7+6], 'Week_day'] = '7-Sun'
Demand = pd.merge(Demand, Location_org, how='left', on='Store')
Demand
```

```python
Duration = pd.read_csv('data/WoolworthsDurations.csv')
Duration.rename(columns={'Unnamed: 0': 'From'}, inplace=True)
Duration = Duration.melt(id_vars=['From'], var_name='To', value_name='Time (Sec)',)
Duration['Time (Sec)'] = Duration['Time (Sec)'] / 3600
Duration.rename(columns={'Time (Sec)': 'Time (Hours)'}, inplace=True)
Duration
```

```python
Demand_org = Demand_org.transpose()
header = Demand_org.iloc[0]
Demand_org = pd.DataFrame(Demand_org.values[1:], columns=header)
Demand_org
```

```python
list_without_6_7 = [i for i in range(28) if ((i+1) % 7 !=0 and (i+2) % 7 != 0)]
list_with_6 = [i for i in range(28) if ((i+2) % 7 == 0)]
list_with_7 = [i for i in range(28) if ((i+1) % 7 == 0)]
store_list_15 = []
store_list_6 = []
store_list_7 = []
store_list_15m = []
store_list_6m = []
store_list_7m = []


z_value = 1.41 
# 92 % of time customer will not have a stockout, assuming the pallet delivered correspond to demand of the stock.
# 5-10 industry standard, 2-3 best in class


for store in Demand_org.columns:
    store_list_15.append(Demand_org.iloc[list_without_6_7][store].std(axis=0) * z_value)
    store_list_6.append(Demand_org.iloc[list_with_6][store].std(axis=0) * z_value)
    store_list_7.append(Demand_org.iloc[list_with_7][store].std(axis=0) * z_value)
    store_list_15m.append(Demand_org.iloc[list_without_6_7][store].mean())
    store_list_6m.append(Demand_org.iloc[list_with_6][store].mean())
    store_list_7m.append(Demand_org.iloc[list_with_7][store].mean())

mon_fri_sd_dict = dict(zip(Demand_org.columns.values, store_list_15))
sat_sd_dict = dict(zip(Demand_org.columns.values, store_list_6))
sun_sd_dict = dict(zip(Demand_org.columns.values, store_list_7))

mon_fri_mean_dict = dict(zip(Demand_org.columns.values, store_list_15m))
sat_mean_dict = dict(zip(Demand_org.columns.values, store_list_6m))
sun_mean_dict = dict(zip(Demand_org.columns.values, store_list_7m))
```

```python
demand_store_15 = []
demand_store_6 = []
demand_store_7 = []
for store in Demand_org.columns:
    demand_store_15.append(mon_fri_sd_dict[store] + mon_fri_mean_dict[store])
    demand_store_6.append(sat_sd_dict[store] + sat_mean_dict[store])
    demand_store_7.append(sun_sd_dict[store] + sun_mean_dict[store])

demand_15_dict = dict(zip(Demand_org.columns.values, demand_store_15))
demand_6_dict = dict(zip(Demand_org.columns.values, demand_store_6))
demand_7_dict = dict(zip(Demand_org.columns.values, demand_store_7))

demand_15_df = pd.DataFrame.from_dict(demand_15_dict, orient='index')
demand_6_df = pd.DataFrame.from_dict(demand_6_dict, orient='index')
demand_7_df = pd.DataFrame.from_dict(demand_7_dict, orient='index')

demand_15 = pd.Series(demand_store_15, index = Demand_org.columns.values)
demand_6 = pd.Series(demand_store_6, index = Demand_org.columns.values)
demand_7 = pd.Series(demand_store_7, index = Demand_org.columns.values)

Demand_Estimation = pd.DataFrame({'Mon-Fri': demand_15,
                                  'Sat': demand_6,
                                  'Sun': demand_7
                                  })

Demand_Estimation.to_csv('data/Demand_Estimation.csv', index=True)
Demand_Estimation
```
